```python
import vertexai
from vertexai.language_models import TextGenerationModel
```

```python
vertexai.init(project="370434740007", location="us-central1")
parameters = {
    "candidate_count": 2,
    "max_output_tokens": 860,
    "temperature": 0.2,
    "top_p": 0.8,
    "top_k": 40
}
model = TextGenerationModel.from_pretrained("text-bison@001")
model = model.get_tuned_model("projects/370434740007/locations/us-central1/models/6082302611787087872")
```

```python
import io
import os
import time
with open(os.path.join('/content/drive/My Drive', 'text.txt'), 'r') as f:
  lines = f.readlines()
new_context = """"""
library = []
# Append each line inside an already defined string
context = """Clasifica el texto únicamente en una de las siguientes categorías:
- PADRES SOLTEROS
- JOVENES
- ADULTOS
- LGTB
- PADRES
- PADRES CASADOS
Si el texto no se ajusta a ninguna categoría, clasifícala de la siguiente manera:
- NO DETERMINABLE
Texto:"""
for line in lines:
  new_context = context + line
  response = model.predict(
    new_context,
    **parameters
  )
  library.append(response.text)
  time.sleep(2)

# Export the library as a text file
with open("library.txt", "w") as f:
  f.writelines([str(response) for response in library])

print(library)

with open('/content/drive/My Drive/output.txt', 'w') as file:
    for item in library:
        file.write("%s\n" % item)
```
